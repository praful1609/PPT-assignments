### Question 1
Given three integer arrays arr1, arr2 and arr3 **sorted** in **strictly increasing** order, return a sorted array of **only** the integers that appeared in **all** three arrays.

**Example 1:**

Input: arr1 = [1,2,3,4,5], arr2 = [1,2,5,7,9], arr3 = [1,3,4,5,8]

Output: [1,5]

**Explanation:** Only 1 and 5 appeared in the three arrays.


```python
def findCommonElements(arr1, arr2, arr3):
    ptr1, ptr2, ptr3 = 0, 0, 0
    result = []

    while ptr1 < len(arr1) and ptr2 < len(arr2) and ptr3 < len(arr3):
        if arr1[ptr1] == arr2[ptr2] == arr3[ptr3]:
            result.append(arr1[ptr1])
            ptr1 += 1
            ptr2 += 1
            ptr3 += 1
        elif arr1[ptr1] < arr2[ptr2]:
            ptr1 += 1
        elif arr2[ptr2] < arr3[ptr3]:
            ptr2 += 1
        else:
            ptr3 += 1

    return result


arr1 = [1,2,3,4,5]
arr2 = [1,2,5,7,9]
arr3 = [1,3,4,5,8]

findCommonElements(arr1, arr2, arr3)
```

### Explanation

1. The function `findCommonElements` takes three sorted arrays, `arr1`, `arr2`, and `arr3`, as input.

2. Three pointers, `ptr1`, `ptr2`, and `ptr3`, are initialized to 0, representing the starting index of each array.

3. An empty array, `result`, is created to store the common elements.

4. The while loop continues as long as all three pointers (`ptr1`, `ptr2`, and `ptr3`) are within the bounds of their respective arrays.

5. Inside the while loop, the code checks if the elements at the current positions of `ptr1`, `ptr2`, and `ptr3` are equal. If they are equal, it means that the current element is present in all three arrays.

6. In that case, the element is appended to the `result` array, and all three pointers are incremented to move to the next elements in each array.

7. If the elements at `ptr1`, `ptr2`, and `ptr3` are not equal, the code checks which element is the smallest among the three. The pointer corresponding to the smallest element is incremented to move to the next element in that particular array.

8. This process continues until any of the pointers reaches the end of its respective array.

9. Finally, the `result` array, which contains the common elements, is returned as the output of the function.

In the given example, the code compares the three arrays [1, 2, 3, 4, 5], [1, 2, 5, 7, 9], and [1, 3, 4, 5, 8].

The code iterates through the arrays using the pointers `ptr1`, `ptr2`, and `ptr3`. It identifies that 1 and 5 are the only elements present in all three arrays, and these elements are appended to the `result` array.

Finally, the `result` array [1, 5] is returned as the output, which represents the sorted array of integers that appeared in all three input arrays.

### Question 2

Given two **0-indexed** integer arrays nums1 and nums2, return *a list* answer *of size* 2 *where:*

- answer[0] *is a list of all **distinct** integers in* nums1 *which are **not** present in* nums2*.*
- answer[1] *is a list of all **distinct** integers in* nums2 *which are **not** present in* nums1.

**Note** that the integers in the lists may be returned in **any** order.

**Example 1:**

**Input:** nums1 = [1,2,3], nums2 = [2,4,6]

**Output:** [[1,3],[4,6]]

**Explanation:**

For nums1, nums1[1] = 2 is present at index 0 of nums2, whereas nums1[0] = 1 and nums1[2] = 3 are not present in nums2. Therefore, answer[0] = [1,3].

For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas nums2[1] = 4 and nums2[2] = 6 are not present in nums2. Therefore, answer[1] = [4,6].


```python
def findDistinct(nums1, nums2):
    diff_nums1 = list(set(nums1) - set(nums2))
    diff_nums2 = list(set(nums2) - set(nums1))
    answer = [diff_nums1, diff_nums2]
    return answer

nums1 = [1,2,3]
nums2 = [2,4,6]
findDistinct(nums1, nums2)
```

1. The function `findDistinct` takes two lists, `nums1` and `nums2`, as input.

2. The code converts `nums1` and `nums2` into sets using the `set()` function. Converting them to sets removes any duplicate elements and transforms the lists into unordered collections of distinct elements.

3. The set difference operation (`-`) is used to find the distinct elements in `nums1` that are not present in `nums2`. The result is stored in the variable `diff_nums1`.

4. Similarly, the set difference operation is used to find the distinct elements in `nums2` that are not present in `nums1`. The result is stored in the variable `diff_nums2`.

5. To obtain the final output format, the code converts `diff_nums1` and `diff_nums2` back into lists using the `list()` function.

6. The code creates a list called `answer`, where `answer[0]` represents the distinct integers in `nums1` not present in `nums2`, and `answer[1]` represents the distinct integers in `nums2` not present in `nums1`. The lists `diff_nums1` and `diff_nums2` are assigned to the respective elements of `answer`.

7. Finally, the `answer` list is returned as the output of the function.

In the given example, the code compares the two arrays `[1, 2, 3]` and `[2, 4, 6]`.

After converting `nums1` and `nums2` into sets, the code finds the set difference between them. It identifies that 1 and 3 are the distinct elements in `nums1` that are not present in `nums2`. Similarly, 4 and 6 are the distinct elements in `nums2` that are not present in `nums1`.

The code converts `diff_nums1` and `diff_nums2` back into lists and assigns them to `answer[0]` and `answer[1]`, respectively.

Finally, the `answer` list `[[1, 3], [4, 6]]` is returned as the output, representing the distinct integers in `nums1` and `nums2` that are not present in each other.

### Question 3
Given a 2D integer array matrix, return *the **transpose** of* matrix.

The **transpose** of a matrix is the matrix flipped over its main diagonal, switching the matrix's row and column indices.

**Example 1:**

Input: matrix = [[1,2,3],[4,5,6],[7,8,9]]

Output: [[1,4,7],[2,5,8],[3,6,9]]


```python
def transposeMatrix(matrix):
    rows = len(matrix)
    columns = len(matrix[0])
    transpose = [[0 for _ in range(rows)] for _ in range(columns)]

    for i in range(rows):
        for j in range(columns):
            transpose[j][i] = matrix[i][j]

    return transpose

matrix = [[1,2,3],[4,5,6],[7,8,9]]
transposeMatrix(matrix)
```

### Explanation
Certainly! Here's a step-by-step explanation of the code:

1. The function `transposeMatrix` takes a 2D integer array `matrix` as input.

2. The code determines the dimensions of the original matrix by getting the number of rows and the number of columns. These values are stored in the variables `rows` and `columns`, respectively.

3. An empty matrix called `transpose` is created using a nested list comprehension. The dimensions of `transpose` are swapped compared to the original matrix. It has `columns` number of rows and `rows` number of columns. Each element of `transpose` is initialized to 0.

4. The nested loops iterate over the rows (`i`) and columns (`j`) of the original matrix.

5. Inside the loops, the code assigns the element at position `matrix[i][j]` to the corresponding position in the transpose matrix, `transpose[j][i]`. This operation swaps the elements from the original matrix to their respective positions in the transpose matrix.

6. After iterating through all the elements of the original matrix, the `transpose` matrix is complete.

7. Finally, the `transpose` matrix is returned as the output of the function.

In the given example, the code takes the matrix `[[1, 2, 3], [4, 5, 6], [7, 8, 9]]`.

The code determines that the original matrix has 3 rows and 3 columns. It creates an empty `transpose` matrix with 3 columns and 3 rows.

The nested loops iterate through each element of the original matrix. The element at position `matrix[i][j]` is assigned to the corresponding position in the `transpose` matrix, `transpose[j][i]`. This effectively flips the rows and columns of the original matrix.

After processing all the elements, the `transpose` matrix `[[1, 4, 7], [2, 5, 8], [3, 6, 9]]` is obtained and returned as the output, representing the transpose of the original matrix.

### Question 4
Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2), ..., (an, bn) such that the sum of min(ai, bi) for all i is **maximized**. Return *the maximized sum*.

**Example 1:**

Input: nums = [1,4,3,2]

Output: 4

**Explanation:** All possible pairings (ignoring the ordering of elements) are:

1. (1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3

2. (1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3

3. (1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4

So the maximum possible sum is 4.


```python
def arrayPairSum(nums):
    nums.sort()
    max_sum = 0
    for i in range(0, len(nums), 2):
        max_sum += nums[i]
    return max_sum


nums = [1, 4, 3, 2]
arrayPairSum(nums)
```

### Explanation
1. The function `arrayPairSum` takes an integer array `nums` as input.

2. The code sorts the `nums` array in ascending order using the `sort()` method. Sorting the array is necessary to pair the smallest numbers with each other and the larger numbers with each other.

3. The variable `max_sum` is initialized to 0. This variable will store the maximum possible sum of the minimum elements in each pair.

4. The code uses a `for` loop to iterate over the sorted `nums` array. The loop starts from index 0 and increments by 2 in each iteration. This step size of 2 ensures that we only consider the smallest numbers in each pair.

5. Inside the loop, the value of the element at the current index, `nums[i]`, is added to `max_sum`. Since the array is sorted in ascending order, `nums[i]` represents the minimum element in the pair.

6. After iterating through all the elements, the `max_sum` variable contains the maximum possible sum of the minimum elements in each pair.

7. Finally, the `max_sum` is returned as the output of the function.

In the given example, the code takes the input array `[1, 4, 3, 2]`.

The code sorts the array to `[1, 2, 3, 4]`. The possible pairings are (1, 2) and (3, 4). The minimum elements in each pair are 1 and 3, respectively.

The code iterates over the sorted array with a step size of 2. It adds the minimum elements (`nums[i]`) to the `max_sum` variable. In this case, `max_sum` is updated to 1 + 3 = 4.

Finally, the code returns the output `4`, representing the maximized sum of the minimum elements in each pair.

### Question 5
You have n coins and you want to build a staircase with these coins. The staircase consists of k rows where the ith row has exactly i coins. The last row of the staircase **may be** incomplete.

Given the integer n, return *the number of **complete rows** of the staircase you will build*.


```python
def arrangeCoins(n):
    k = 1
    while n >= k:
        n -= k
        k += 1
    return k - 1

n = 8
arrangeCoins(n)

```

### Explanation
Certainly! Here's a step-by-step explanation of the corrected code:

1. The function `arrangeCoins` takes an integer `n` as input.

2. The variable `k` is initialized to 1. This variable represents the number of rows in the staircase.

3. The code enters a `while` loop, which continues as long as `n` is greater than or equal to `k`.

4. Inside the loop, `k` is subtracted from `n`. This operation reduces the number of remaining coins (`n`) by the number of coins in the current row (`k`).

5. The variable `k` is incremented by 1. This moves the loop to the next iteration, representing the next row of the staircase.

6. Once the loop exits, the value of `k` is one greater than the number of complete rows in the staircase.

7. To obtain the correct count of complete rows, `k` is decremented by 1 before being returned as the final output.



### Question 6
Given an integer array nums sorted in **non-decreasing** order, return *an array of **the squares of each number** sorted in non-decreasing order*.

**Example 1:**

Input: nums = [-4,-1,0,3,10]

Output: [0,1,9,16,100]

**Explanation:** After squaring, the array becomes [16,1,0,9,100].
After sorting, it becomes [0,1,9,16,100]


```python
def sortedSquares(nums):
    result = []
    for num in nums:
        result.append(num * num)
    result.sort()
    return result


nums = [-4, -1, 0, 3, 10]
sortedSquares(nums)

```

### Explanation
Certainly! Here's a step-by-step explanation of the code:

1. The function `sortedSquares` takes an integer array `nums` as input.

2. An empty array `result` is initialized to store the squared numbers.

3. The code iterates over each element `num` in the input array `nums`.

4. Inside the loop, the code computes the square of `num` by multiplying it with itself (`num * num`).

5. The squared value is then appended to the `result` array using the `append()` method.

6. After iterating through all the elements in `nums`, the `result` array contains the squared values in the same order as the original array.

7. The `result` array is then sorted in non-decreasing order using the `sort()` method.

8. Finally, the sorted `result` array is returned as the output of the function.


### Question 7
You are given an m x n matrix M initialized with all 0's and an array of operations ops, where ops[i] = [ai, bi] means M[x][y] should be incremented by one for all 0 <= x < ai and 0 <= y < bi.

Count and return *the number of maximum integers in the matrix after performing all the operations*


```python
def maxCount(m, n, ops):
    min_row = m
    min_col = n

    for op in ops:
        min_row = min(min_row, op[0])
        min_col = min(min_col, op[1])

    return min_row * min_col

m = 3
n = 3
ops = [[2, 2], [3, 3]]
maxCount(m, n, ops)

```

### Explanation
Certainly! Here's a step-by-step explanation of the code:

1. The function `maxCount` takes three parameters: `m` and `n` representing the dimensions of the matrix `M`, and `ops` which is an array of operations.

2. The variables `min_row` and `min_col` are initialized as `m` and `n` respectively. These variables will track the minimum row and column values affected by the operations.

3. The code iterates over each operation `[ai, bi]` in the `ops` array.

4. Inside the loop, the code updates `min_row` as the minimum value between the current value of `min_row` and `ai`. This ensures that `min_row` will always store the minimum row value affected by the operations.

5. Similarly, the code updates `min_col` as the minimum value between the current value of `min_col` and `bi`. This ensures that `min_col` will always store the minimum column value affected by the operations.

6. After iterating through all the operations, `min_row` will contain the minimum row value affected, and `min_col` will contain the minimum column value affected.

7. The code returns the product of `min_row` and `min_col` as the count of maximum integers. Since the operations increment all the values within the affected area, the maximum integer value will be present in all cells of that area.


### Question 8

Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].

*Return the array in the form* [x1,y1,x2,y2,...,xn,yn].

**Example 1:**

**Input:** nums = [2,5,1,3,4,7], n = 3

**Output:** [2,3,5,4,1,7]

**Explanation:** Since x1=2, x2=5, x3=1, y1=3, y2=4, y3=7 then the answer is [2,3,5,4,1,7].


```python
def rearrangeArray(nums, n):
    result = []
    for i in range(n):
        result.append(nums[i])
        result.append(nums[i+n])
    return result

nums = [2, 5, 1, 3, 4, 7]
n = 3
rearrangeArray(nums, n)

```

### Explanation
1. The function `rearrangeArray` takes two parameters: `nums`, which is the input array, and `n`, which represents half the length of the array.

2. An empty array `result` is initialized to store the rearranged elements.

3. The code iterates over `i` from 0 to `n-1`, since we need to rearrange the first `n` elements.

4. Inside the loop, the code performs two actions for each iteration:

   - It appends `nums[i]` to the `result` array. This corresponds to taking the `i`th element from the first half of `nums` (xi).
   - It appends `nums[i+n]` to the `result` array. This corresponds to taking the `i`th element from the second half of `nums` (yi).

   By doing this, the code interleaves the elements from the first half and the second half of `nums` in the desired form.

5. After iterating through all the elements, the `result` array contains the rearranged elements.

6. The `result` array is then returned as the output of the function.
